```python
# calculate the corralation between market_value and fpl_value
corr_value=df['market_value'].corr(df['fpl_value'])
corr_value
```

```python
# the corrlation between market value and fpl value is close to 1 it means  
# there is a nearly postve relationship, so who have a high market value 
# will have a high fpl value
```
